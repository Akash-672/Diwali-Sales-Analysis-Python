**Exploratory Data Anlaysis (EDA)**

```python
df.columns
```

```python
nam = sns.countplot(x='Gender',data=df)

for bars in nam.containers:
    nam.bar_label(bars)

#This is tell us which gender (M/F) made the most purchase
```

```python
gen_sales = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Gender',y='Amount',data=gen_sales)

#same as above but my amount.
```

From the above graphs we can say that most of the buyers are females and by purchasing amount also

**Age**

```python
MG = sns.countplot(x='Age Group',data=df,hue = 'Gender')

for bars in MG.containers:
    MG.bar_label(bars)

#for thing will gives the numbers above the bars that it
```

```python
sales = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Age Group',y='Amount',data=sales)
```

From the above graph we can that age between 26-35 and most of them are females

**State**

```python
df.columns
```

```python
state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(18,5)})
sns.barplot(data = state, x='State',y='Orders')

#orders by state
```

```python
state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(18,5)}) #this will tell the size of the plot
sns.barplot(data = state, x='State',y='Amount')

#amount by state
```

we can see from above graph we can see that most of the sales/amount are from uttar pradesh, maharashtra , karnataka respectiely

**Marital_Status**

```python
FG = sns.countplot(data=df, x='Marital_Status')

sns.set(rc={'figure.figsize':(5,5)})
for bars in FG.containers:
    FG.bar_label(bars)
```

```python
sales = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales, x='Marital_Status',y='Amount', hue='Gender')

#this will give me gender
```

From the above graph we can say that most of the purchase are made by married women

**Occupation**

```python
sns.set(rc={'figure.figsize':(20,5)})
FG = sns.countplot(data=df, x='Occupation')

for bars in FG.containers:
    FG.bar_label(bars)
```

```python
No = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = No, x='Occupation',y='Amount')
# this will tell by occupation purchase in desc
```

From the above graph we can that IT, Healthcare , Aviation occupations customers purchased the most

**Product by Category**

```python
sns.set(rc={'figure.figsize':(25,5)})
FG = sns.countplot(data=df, x='Product_Category')

for bars in FG.containers:
    FG.bar_label(bars)
#
```

```python
HI = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(25,10)})
sns.barplot(data = HI, x = 'Product_Category',y = 'Amount')
```

From the above graphs we can say that most of purchased categories are Food , clothing , electronics

**Conclusion**

From the Above EDA analysis. Married women of age group between 26-35 yrs from UP, maharashtra, karnataka working in IT, Healthcare, Aviation are most likely buy products from Food , cloths and electronics category.
